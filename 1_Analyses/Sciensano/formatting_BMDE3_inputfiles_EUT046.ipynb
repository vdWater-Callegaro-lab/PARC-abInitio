```python
import pandas as pd
import numpy as np
```

```python
EUT046_normalized = pd.read_csv('../datasets/EUT046_RODAFpreprocessed_normalizedcounts.csv', index_col='Unnamed: 0')
```

```python
EUT046_normalized.replace(0, 1, inplace=True)
```

```python
EUT046_log2normalized = np.log2(EUT046_normalized)
```

```python
EUT046_log2normalized
```

```python
EUT046_normalized.reset_index(drop=False, inplace=True)
```

```python
EUT046_log2normalized.reset_index(drop=False, inplace=True)
```

```python
EUT046_normalized = EUT046_normalized.rename(columns={'index': '0_0_Dose'})
```

```python
EUT046_log2normalized = EUT046_log2normalized.rename(columns={'index': '0_0_Dose'})
```

```python
EUT046_log2normalized
```

```python
EUT046_normalized = pd.concat([EUT046_normalized,pd.DataFrame({col:col.split('_')[2] for col in EUT046_normalized.columns}, index=[1])])
```

```python
EUT046_log2normalized = pd.concat([EUT046_log2normalized,pd.DataFrame({col:col.split('_')[2] for col in EUT046_log2normalized.columns}, index=[1])])
```

```python
conc = {'DMEM':0, 'Cis9':0.01, 'Cis8':0.5, 'Cis7':1, 'Cis6':2.5, 'Cis5':5, 'Cis4':10, 'Cis3':20, 'Cis2':30, 'Cis1':50}
```

```python
EUT046_normalized = EUT046_normalized.replace(conc)
```

```python
EUT046_log2normalized = EUT046_log2normalized.replace(conc)
```

```python
derniere_ligne = EUT046_normalized.iloc[-1]
EUT046_normalized = pd.concat([derniere_ligne.to_frame().T, EUT046_normalized.iloc[:-1]])
```

```python
derniere_ligne = EUT046_log2normalized.iloc[-1]
EUT046_log2normalized = pd.concat([derniere_ligne.to_frame().T, EUT046_log2normalized.iloc[:-1]])
```

```python
EUT046_normalized.reset_index(drop=True, inplace=True)
```

```python
EUT046_log2normalized.reset_index(drop=True, inplace=True)
```

```python
EUT046_log2normalized
```

```python
timepoint_l = ['4h', '8h', '16h', '24h', '48h', '72h']
timepoint_d = dict()
for timepoint in timepoint_l:
    key1 = f'EUT046_RODAFpreprocessed_normalized{timepoint}'
    value1 = EUT046_normalized.filter(regex=f'_{timepoint}_|Dose')
    timepoint_d[key1] = value1

for k,v in timepoint_d.items():  
    
    v.to_csv(f'../datasets/EUT046_RODAF_preproccessed_forBMD/{k}_forBMDExpress.txt', sep='\t', index=False)
```
