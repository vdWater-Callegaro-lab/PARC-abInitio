```python
libsize_df = pd.DataFrame({'SAMPLE_ID':list(EUT046.columns)[1:], 'LIB_SIZE':[sum(EUT046[c]) for c in list(EUT046.columns)[1:]]})
```

```python
EUT046_manifest.rename(columns={'Probe_ID':'PROBE_ID'}, inplace=True)
```

```python
EUT046.rename(columns={'Column1':'PROBE_ID'}, inplace=True)
```

```python
def calculate_cpm_for_dataframe(df, sample_sizes_in_millions):

    cpm_df = pd.DataFrame()

    for condition, sample_size_in_millions in zip(list(df.columns)[1:], sample_sizes_in_millions):
        cpm = (df[condition] / sample_size_in_millions)
        cpm_df[condition + '_CPM'] = cpm * 1000000
        
    cpm_df = pd.concat([df['PROBE_ID'], cpm_df], axis=1)
    return cpm_df



    
```

```python
EUT046_CPM = calculate_cpm_for_dataframe(EUT046, list(libsize_df.LIB_SIZE))
```

```python
def assign_boolean(cpm_value):
    if cpm_value >= 1:
        return 1
    else:       
        return 0
```

```python
EUT046_bool = EUT046_CPM.iloc[:,1:].applymap(assign_boolean)
```

```python
EUT046_bool = pd.concat([EUT046.iloc[:,0], EUT046_bool], axis=1)
```

```python
conditions = list(set([e.split('_Rep')[0] for e in EUT046_bool.iloc[:,1:].columns]))
```

```python
def check_75_percent(row):
    # Compter le nombre de valeurs supérieures à 1 pour chaque ligne (gène)
    count_above_one = sum(row[:] >= 1)
    # Vérifier si 75% ou plus des valeurs sont supérieures à 1
    return count_above_one / len(row) >= 0.75
```

```python
filtered_pd = EUT046_bool['PROBE_ID']
for condition in conditions:
    filtered_columns = EUT046_bool.iloc[:,1:].filter(regex=re.compile(condition))
    print(len(filtered_columns.columns))
    filtered_columns[condition] =  filtered_columns.apply(check_75_percent, axis=1)
    filtered_pd = pd.concat([filtered_pd, filtered_columns[condition]], axis=1)
```

```python
preprocessed_EUT046 = filtered_pd[filtered_pd.iloc[:,1:].any(axis=1)]
```

```python
preprocessed_EUT046_counts = pd.merge(preprocessed_EUT046[['PROBE_ID']], EUT046, how='left', on='PROBE_ID')
```

```python
preprocessed_EUT046_to_sum = pd.merge(preprocessed_EUT046_counts, EUT046_manifest[['PROBE_ID', 'gene_symbol']], how='left', on='PROBE_ID')
```

```python
preprocessed_EUT046_sumcounts = preprocessed_EUT046_to_sum.groupby('gene_symbol', as_index=False).sum()
```
