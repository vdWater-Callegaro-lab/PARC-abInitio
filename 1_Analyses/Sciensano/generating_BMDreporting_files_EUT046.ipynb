```python
pathway = pd.read_csv('../data/input/EUT046_sciensano/pathway_overview.txt', sep='\t')
```

```python
pathway[pathway.pathway == 'TGxDDI64']
```

```python
Analysis = []
pathwayname = []
amountgenes = []
medians = []
mediansBMDL = []
mediansBMDU = []
i = 0
for e in set(pathway.pathway):
    res_d_pathway = dict()
    #print(e)
    i += 1
    #print(i)
    for res in res_d.keys():
        res_pathway = pd.merge(pathway[pathway.pathway == e], res_d[res], how='inner', on='entrez_id')
        if len(res_pathway[['Best BMD']]) > 0: 
            Analysis.append(res)
            pathwayname.append(e)
            amountgenes.append(len(res_pathway.entrez_id))
            medians.append(stat.median(res_pathway['Best BMD']))
            mediansBMDL.append(stat.median(res_pathway['Best BMDL']))
            mediansBMDU.append(stat.median(res_pathway['Best BMDU']))
            
median_df = pd.DataFrame({'Analysis':Analysis, 'Pathway.Name':pathwayname, 'Genes.That.Passed.All.Filters':amountgenes, 'Median':medians, 'BMDL.Median':mediansBMDL, 'BMDU.Median':mediansBMDU})
```
